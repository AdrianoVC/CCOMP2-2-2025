```python
def  insert_sort(A):
  for i in range(1,len(A)):
    actual=A[i]
    j=i-1

    while j>=0  and A[j]>actual:
      A[j+1]=A[j]
      j-=1
    A[j+1]=actual

  return A



A=[7,3,9,1,2]
print(insert_sort(A))

```

```python
def quick_sort(A):

  if len(A)<=1:
    return A
  else:
    pivote=A[len(A)//2]
    menores=[]
    iguales=[]
    mayores=[]

    for e in range(len(A)):
      if A[e]< pivote:
        menores.append(A[e])
      elif A[e]> pivote:
        mayores.append(A[e])
      else:
        iguales.append(A[e])
    return quick_sort(menores)+iguales+quick_sort(mayores)


A=[7,3,9,1,2]
print(quick_sort(A))

```
